## 回归概述

回归方法是最古老、使用最广泛的分析方法之一。 

回归的目标是生成一个模型，该模型代表对某些观察到的数据的“最佳拟合（best fit）”。 

通常，模型是描述某种类型的曲线（直线、抛物线等）的函数，由一组参数（例如，斜率和截距）确定。 

“最佳拟合”是指根据我们选择的评估标准，存在一组最优参数。

回归模型试图预测一个变量的值，称为 **因变量**、**响应变量** 或 **标签**，使用其他变量的值，称为 **自变量** 、**解释变量**或**特征**。 单一回归有一个标签用于预测一个特征。 多元回归使用两个或多个特征变量。

在数学形式中，回归的目标是找到预测标签值 $y$ 的某些特征 $X$ 的函数。 这个函数可以写成如下：

$$\hat{y} = f(X)$$

回归中的挑战是**学习**函数 $f(X)$ 以便 $\hat{y}$ 的预测是准确的。

换句话说，我们训练模型以最小化我们预测的 $\hat{y}$ 和已知标签值 $y$ 之间的差异。

事实上，**监督学习**的整个领域都有这个目标

许多机器学习模型，包括一些最新的深度学习方法，都是回归的一种形式。

这些方法经常遇到相同的问题，包括过度拟合和数学上不稳定的拟合方法。

### 线性回归概述

在本实验中，您将使用线性回归模型。

线性回归是回归的一种基本形式。

一旦你对线性回归有了一点了解，你就会对机器学习有很多了解。

线性回归的最简单情况被称为**单一回归**，因为只有一个特征。 函数 $f(X)$ 在模型系数中是**线性的**。

对于单个特征向量 $x$，线性回归方程如下所示：

$$\hat{y} = a \cdot x + b$$

模型系数是 $a$，我们称之为 **斜率** 和 $b$，我们称之为 **截距**。 请注意，这只是一个变量的直线方程。

但是，$a$ 和 $b$ 的最佳值是多少？ 

在线性回归中，选择 $a$ 和 $b$ 以最小化预测与已知标签之间的平方误差。 此数量称为**误差平方和**或 SSE。 对于 $n$ **训练案例**，SSE 计算如下：

$$MSE = \sum_{i=1}^n \big( f(x_i) - y_i \big)^2\\
= \sum_{i=1}^n \big( \hat{y}_i - y_i \big)^2\\
= \sum_{i=1}^n \big( a \cdot x_i + b - y_i \big)^2$$

最小化 SSE 的回归方法被称为**最小二乘法**。

### 执行第一个线性回归示例

考虑到这一点理论，您现在将训练和评估线性回归模型。 在这种情况下，您将使用模拟数据，这意味着您可以将计算结果与数据的已知属性进行比较。

第一步，执行下面单元格中的代码以加载运行本笔记本其余部分所需的包。

```python
## Import packages
import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt
import sklearn.model_selection as ms
import sklearn.metrics as sklm
from sklearn import preprocessing
from sklearn import linear_model
import scipy.stats as ss
import seaborn as sns
import math

# make plots appear inline in the notebook
%matplotlib inline 
```

下面单元格中的代码模拟数据并绘制结果。 数据具有以下属性：

- `x` 变量在 0.0 和 10.0 之间均匀分布。
- `y` 变量等于 `x` 变量加上正态分布的随机分量。 因此，对于未标定的数据，斜率系数应为 1.0，截距应为 0.0。

执行此代码并检查结果。

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
nr.seed(34567)
x = np.arange(start = 0.0, stop = 10.0, step = 0.1)
y = np.add(x, nr.normal(scale = 1.0, size = x.shape[0]))

sns.regplot(x, y, fit_reg = False)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data for regression')
```

正如预期的那样，这些数据遵循直线趋势。 但是，由于添加了正态分布的噪声，这些数据会有些分散。

#### 拆分数据集

在执行任何类型的机器学习时，都需要良好的数据准备以确保良好的模型性能。 数据准备不当通常是机器学习模型性能不佳的根源。

准备这些数据的第一步是创建**独立采样****训练数据集**和**测试数据集**。

在大多数情况下，还将使用独立采样的**评估数据集**。

在这种情况下，不会执行任何模型改进或比较，因此不需要这个额外的步骤。

如果使用相同的数据来训练和测试机器学习模型，则该模型很可能只是在学习训练数据。

用技术术语来说，可以说训练和测试过程之间存在**信息泄漏**。 在这种情况下，模型可能无法很好地**泛化**。

当出现以前从未遇到过的新案例时，泛化良好的模型会产生一致的结果。 相反，泛化能力差的模型在处理新案例时可能会给出意想不到的结果。

数据的随机子样本是使用称为**伯努利抽样**的过程创建的。 伯努利抽样以概率 $p$ 将样本接受到数据子集中。

在这种情况下，给定案例在训练数据集中的概率为 $p$。 然后，一个案例在测试数据集中的概率变为 $1-p$。

`sklearn.model_selection` 模块中的 `train_test_split` 函数执行了所需的伯努利采样。 `train_test_split` 函数对包含特征和标签值的数组的索引进行采样。 下面单元格中的代码执行此拆分。 执行这段代码。

```python
## Randomly sample cases to create independent training and test data
nr.seed(9988)
indx = range(len(x))
indx = ms.train_test_split(indx, test_size = 50)
x_train = np.ravel(x[indx[0]])
y_train = np.ravel(y[indx[0]])
x_test = np.ravel(x[indx[1]])
y_test = np.ravel(y[indx[1]])
```

#### 缩放数字特征

现在数据集已拆分，必须重新缩放数字特征列。 数字特征的重新缩放非常重要。 特征的数值范围不应决定该特征在多大程度上决定了机器学习模型的训练。

例如，考虑一个具有两个特征的数据集，年龄（通常以几十为单位）和收入（通常以数万或数十万为单位）。 没有理由相信收入在某些模型中比年龄更重要，仅仅是因为它的值范围更大。 为了防止这个问题，数字特征被缩放到相同的范围。

有许多可能的缩放方法。 一种简单的方法称为 **Min-Max** 归一化。 使用以下公式将数据缩放到 $\{ 0,1 \}$ 范围内：

$$x\_scaled_i = \frac{(x_i - Min(x))}{(Max(x) - Min(x))}$$

在哪里，
$x_i $是第i个样本值，
$Min(X)$是所有样本的最小值，
$Max(X)$ 是所有样本的最大值。

一般来说，最小-最大归一化对于被缩放的值具有复杂分布的情况是一个不错的选择。 例如，具有多种模式分布的变量可能是 Min-Max 归一化的良好候选者。 请注意，一些异常值的存在可能会通过给出不具代表性的 $Min(X)$ 或 $Max(X)$ 值来扭曲结果。


在本实验中，您将使用 **Z-Score** 归一化。 Z-Score 归一化对变量进行变换，使其具有零均值和单位标准差（或方差）。 Z-Score 归一化使用以下公式执行：

$$x\_scaled_i = \frac{\big(x_i - \mu \big)}{\sigma}$$

在哪里，
$\mu $ 是变量 $X$ 的均值，
$\sigma$ 是变量 $X$ 的标准差。

下面单元格中的代码使用了 `sklearn.preprocessing` 包中的 `StandardScaler` 函数。 此函数计算训练数据的比例系数。 然后使用“transform”方法将生成的转换应用于训练和测试数据。

请注意，缩放变换仅在训练数据上计算。 缩放变换应该总是在训练数据上计算，而不是测试或评估数据。

通常，数字标签不需要缩放。 但是，可能需要其他转换。

执行下面单元格中的代码，将 Z-Score 转换应用于训练和测试特征。

```python
## Scale the feature, being sure to use the scale of the training
## data not the test data. 
scaler = preprocessing.StandardScaler().fit(x_train.reshape(-1,1))
x_train = scaler.transform(x_train.reshape(-1,1))     
y_train = scaler.transform(y_train.reshape(-1,1))     
```

## 训练回归模型

准备好数据后，就可以训练回归模型了。 这是通过 `sklearn.linear_model` 包完成的。 大多数 scikit-learn 模型的训练步骤与此处使用的相同：

1. 模型对象被实例化。 可以在实例化时执行其他模型规范。
2. 使用特征和标签的 numpy 数组拟合模型。 在这种情况下，只有一个特征，因此使用“reshape”方法创建一个正确维度的数组。

您可以点击此链接查找其他 [关于使用 scikit-learn 的线性回归模型的文档](http://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares)。

执行下面单元格中的代码以实例化和拟合模型。

```python
## define and fit the linear regression model
lin_mod = linear_model.LinearRegression()
lin_mod.fit(x_train.reshape(-1,1), y_train)
```

作为此模型的第一次验证，您可以打印模型系数。 这些系数是模型对象的属性。 执行下面单元格中的代码并检查结果。

```python
print(lin_mod.intercept_)
print(lin_mod.coef_)
```

这些系数接近模拟中使用的值 $0.0$ 和 $1.0$。

接下来，您将绘制根据训练特征计算的预测值。 

将 “predict” 方法应用于具有训练数据的模型。 然后显示原始标签值的图和预测值或 **scores** 的线。 执行代码并检查结果。

```python
def plot_regression(x, y_score, y):
    ## Plot the result
    sns.regplot(x, y, fit_reg=False)
    plt.plot(x, y_score, c = 'red')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Fit of model to test data')

y_score = lin_mod.predict(x_test.reshape(-1,1)) 

plot_regression(x_test, y_score, y_test)
```

红线似乎与数据非常吻合。 得分值和残差之间的误差似乎很小。 但是，需要对模型性能进行客观评估。

## 评估模型性能

训练好模型后，就可以评估性能了。 这是使用测试数据集完成的，因此模型训练没有信息泄漏。

作为第一步，计算一组性能指标。 有许多可能的指标可用于评估回归模型。 通常，这些指标是**残差**的函数，或者预测值或分数与实际标签值之间的差异：

$$r_i = f(x_i) - y_i = \hat{y}_i - y_i$$

在本实验中，您将使用一些更常见的指标：

- **均方误差**或 MSE，
$$MSE = \frac{1}{N} \sum_{i=1}^N (f(x_i) - y_i)^2$$

均方误差与残差的方差相同（略有偏差）。 回想一下，这个指标是一个线性回归最小化的指标。 请注意，均方误差以标签值的平方为单位。

- **均方根误差**或 RMSE，
$$RMSE = \sqrt{ \frac{1}{N} \sum_{i=1}^N (f(x_i) - y_i)^2}$$

均方根误差与残差的标准差相同（同样，有轻微偏差）。 均方根误差与标签值的单位相同。

- **平均绝对误差**或 MAE，
$$MAE = \frac{1}{N} \sum_{i=1}^N |f(x_i) - y_i|$$
其中 $||$ 是绝对值运算符。

与均方根误差的解释相似。 您可能会发现此度量更直观，因为它只是残差大小的平均值。

- **中位数绝对误差/Median absolute error**，

$$Median\ Absolute\ Error = Median \big( \sum_{i=1}^N |f(x_i) - y_i| \big)$$   

中值绝对误差是绝对残差位置参数的稳健度量。 如果此度量与平均绝对误差显着不同，则残差中可能存在异常值。

- **R 平方或 $R^2$**，也称为 **决定系数**，
$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$
在哪里，
$SS_{res} = \sum_{i=1}^N r_i^2$，或残差平方和，
$SS_{tot} = \sum_{i=1}^N y_i^2$，或者标签值的平方和。

换句话说，$R^2$ 是原始标签值和残差之间的平方值之和减少的量度。 如果模型没有减少标签的平方和（一个无用的模型！），$R^2 = 0$。 另一方面，如果模型与数据完全吻合，那么所有 $r_i = 0$，则 $R^2 = 1$。

- **调整后的 R 平方或 $R^2_{adj}$** 是针对模型中的自由度调整的 $R^2$，
$$R^2_{adj} = 1 - \frac{var(r)}{var(y)} = 1 - \frac{\frac{SS_{res}}{(n - p -1)}}{ \frac{SS_{tot}}{(n-1)}}$$
在哪里，
$var(r) = $残差的方差，
$var(y) = $标签的方差，
$n = $样本或案例的数量，
$p = $ 模型参数的数量。

$R^2_{adj}$的解释与$R^2$相同。 在许多情况下，差别不大。 但是，如果参数数量相对于案例数量而言非常重要，则 $R^2$ 将对模型性能给出过于乐观的度量。 一般来说，$R^2_{adj}$ 和 $R^2$ 之间的差异随着案例数量 $n$ 的增加而变得不那么显着。 然而，即使对于“大数据”模型，如果存在大量模型参数，也会存在显着差异。

****
**注意：** 是否有可能获得超出 $\{ 0,1 \}$ 范围的 $R^2$ 值？ 通常没有。 但也有例外。

$R^2$ 只能在退化情况下大于 $1$。 例如，如果所有标签值都相同。 但是，在这种情况下，您不需要模型来预测标签！

如果您发现您的模型给出的 $R^2$ 小于 $0$ 怎么办？ 这可能意味着什么？ 这总是意味着您的代码中存在错误，并且您的模型的残差比原始标签具有更大的分散性！
****

下面单元格中的代码使用 `sklearn.metrics` 包中的函数来计算一些常用指标。 `sklearn.metrics` 中没有 $R^2_{adj}$ 的函数，但是自由度的调整很容易计算。

您可以点击此链接查找其他 [scikit-learn 中内置的线性回归 merics 文档](http://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics)。

执行下面单元格中的代码，检查结果，并回答课程页面上的**问题 1**。

```python
def print_metrics(y_true, y_predicted, n_parameters):
    ## First compute R^2 and the adjusted R^2
    r2 = sklm.r2_score(y_true, y_predicted)
    r2_adj = r2 - (y_true.shape[0] - 1)/(y_true.shape[0] - n_parameters - 1) * (1 - r2)
    
    ## Print the usual metrics and the R^2 values
    print('Mean Square Error      = ' + str(sklm.mean_squared_error(y_true, y_predicted)))
    print('Root Mean Square Error = ' + str(math.sqrt(sklm.mean_squared_error(y_true, y_predicted))))
    print('Mean Absolute Error    = ' + str(sklm.mean_absolute_error(y_true, y_predicted)))
    print('Median Absolute Error  = ' + str(sklm.median_absolute_error(y_true, y_predicted)))
    print('R^2                    = ' + str(r2))
    print('Adjusted R^2           = ' + str(r2_adj))
    
print_metrics(y_test, y_score, 2)    
```

您如何解释这些结果：
- MSE 和 RMSE 符合预期。 模拟数据的标准偏差为 $1.0$。
- MAE 和中值绝对误差具有较小的值并且靠得很近，表明模型拟合良好且残差中几乎没有明显的异常值。
- $R^2_{adj}$ 和 $R^2$ 都非常接近于 1，表明该模型正在做出有用的预测，这比标签值的简单平均值要好得多。

线性回归模型的残差应该近似服从正态分布。 这种情况可以使用图形方法轻松测试，特别是直方图和分位数-分位数正态图。

****
**注意：**一个常见的误解是线性回归模型的特征或标签必须具有正态分布。 不是这种情况！ 相反，模型的残差（误差）应该服从正态分布。
****

下面单元格中的代码绘制了核密度图和回归模型残差的直方图。 执行此代码并检查结果。

```python
def hist_resids(y_test, y_score):
    ## first compute vector of residuals. 
    resids = np.subtract(y_test.reshape(-1,1), y_score.reshape(-1,1))
    ## now make the residual plots
    sns.distplot(resids)
    plt.title('Histogram of residuals')
    plt.xlabel('Residual value')
    plt.ylabel('count')
    
hist_resids(y_test, y_score)    
```

此直方图和核密度图看起来大致正常，但有一些偏差。 总的来说，这些残差对于真实世界的模型来说看起来是合理的。

另一个有用的图是**分位数-分位数正态图**，或 Q-Q 正态图。 此图在水平轴上显示标准正态分布的分位数，在垂直轴上显示残差的分位数。 如果残差完全服从正态分布，则这些点将落在一条直线上。 在实际问题中，您应该期望直线关系是近似的。

执行下面单元格中的代码并检查结果图。

```python
def resid_qq(y_test, y_score):
    ## first compute vector of residuals. 
    resids = np.subtract(y_test.reshape(-1,1), y_score.reshape(-1,1))
    ## now make the residual plots
    ss.probplot(resids.flatten(), plot = plt)
    plt.title('Residuals vs. predicted values')
    plt.xlabel('Quantiles of standard Normal distribution')
    plt.ylabel('Quantiles of residuals')
    
resid_qq(y_test, y_score)    
```

请注意，这些点几乎落在直线上。 这表明残差具有近似正态的分布。

您现在将为该回归模型制作最后一个诊断图，称为**残差图**。 残差与预测值（分数）的关系图显示残差中是否存在结构。 对于理想的回归模型，残差的方差或离差不应随预测值的变化而变化。 有人说，理想的残差图应该看起来像一条“毛毛虫”，与预测值相比没有变化。

此图中任何结构的预测值发生变化都表明模型拟合随预测值而变化。 例如，如果残差随着预测值的增加而增加，则可以说该模型仅能很好地预测较小的标签值。 相反的情况表明只有大标签值才能得到很好的预测。 中间范围的变化表明预测值存在一些非线性变化。 换句话说，在任何这些情况下，模型都没有准确计算预测值。

执行下面单元格中的代码以显示和检查回归模型的残差图。

```python
def resid_plot(y_test, y_score):
    ## first compute vector of residuals. 
    resids = np.subtract(y_test.reshape(-1,1), y_score.reshape(-1,1))
    ## now make the residual plots
    sns.regplot(y_score, resids, fit_reg=False)
    plt.title('Residuals vs. predicted values')
    plt.xlabel('Predicted values')
    plt.ylabel('Residual')
    
resid_plot(y_test, y_score)    
```

这个残差图看起来表现得相当好。 在预测值的范围内，离散度相当恒定。
