```python
auto_prices = pd.read_csv('Automobile price data _Raw_.csv')
auto_prices
```

```python
def clean_auto_data(auto_prices):
    'Function to load the auto price data set from a .csv file' 
    import pandas as pd
    import numpy as np
    
    ## 对名称重新编码
    ## Recode names
    ## fix column names so the '-' character becomes '_'
    cols = auto_prices.columns
    auto_prices.columns = [str.replace('-', '_') for str in cols]
    
    ## 处理缺失数据，移除缺失数据，将 ？看成是缺失数据。
    ## Treat missing values
    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 'horsepower', 'peak_rpm']
    for column in cols:
        auto_prices.loc[auto_prices[column] == '?', column] = np.nan
    auto_prices.dropna(axis = 0, inplace = True)

    ## 处理数据类型，将所有的是值转换为 numeric 类型。
    ## Transform column data type
    ## Convert some columns to numeric values
    for column in cols:
        auto_prices[column] = pd.to_numeric(auto_prices[column])

    return auto_prices

auto_prices = clean_auto_data(auto_prices)

print(auto_prices.columns)
```

### 条形图

作为第一步，您将使用**条形图/bar charts**来检查分类变量的频率分布。 条形图显示每个类别的频率。 

在大多数情况下，类别应该**按频率排序/ordered by frequency**； 上升或下降。 按频率排序类别有助于观众理解。

执行下面单元格中的函数，进行如下处理：

1. 遍历列列表。
2. 图形和轴使用Matplotlib方法定义。
3. 计算类别的计数或频率/The counts or frequencies of the categories are computed。
4. 使用 Pandas 的“plot.bar”方法创建条形图。 请注意，颜色参数设置为蓝色。 无缘无故地默认使用**多种颜色会分散注意力，并且不会增加解释**。
5. 使用 Matplotlib 方法将注释添加到图中。

由于 Pandas 绘图是基于 Matplotlib 构建的，因此始终可以使用此包中的方法添加额外的绘图属性。

执行此代码并检查结果。

```python
def plot_bars(auto_prices, cols):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # 定义绘图区域
        ax = fig.gca() # 定义轴axis    
        counts = auto_prices[col].value_counts() # 找到每个唯一类别的计数
        counts.plot.bar(ax = ax, color = 'blue') # 在计数数据框上使用 plot.bar 方法
        ax.set_title('Number of autos by' + col) # 给一个主标题
        ax.set_xlabel(col) # 设置 x 轴的文本
        ax.set_ylabel('Number of autos')# 为 y 轴设置文本
        plt.show()

plot_cols = ['make', 'body_style', 'num_of_cylinders']
plot_bars(auto_prices, plot_cols)    
```

这些图显示了所绘制的分类变量的广泛频率范围。 这将是建模的问题，因为某些类的成员很少。

### 直方图/Histograms

**直方图/Histograms**与条形图有关。 

条形图显示唯一类别的计数，而直方图显示 **数值变量** 的 bin**内数据值的**数量。 

箱子将变量的值分成相等的段。 直方图的垂直轴显示每个 bin 中数据值的计数。

下面的代码遵循用于条形图创建直方图的相同基本方法。 在这种情况下，使用 Pandas 的“plot.hist”方法。

执行此代码，检查结果，并回答课程页面上的**问题 1**。

```python
def plot_histogram(auto_prices, cols, bins = 10):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis    
        auto_prices[col].plot.hist(ax = ax, bins = bins) # Use the plot.hist method on subset of the data frame
        ax.set_title('Histogram of ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of autos')# Set text for y axis
        plt.show()
        
num_cols = ['curb_weight', 'engine_size', 'city_mpg', 'price']    
plot_histogram(auto_prices, num_cols)
```

## 判断正态分布的三大法宝

### 偏度/Skewness、峰度/Kurtosis

* 偏度/Skewness

偏度衡量随机变量概率分布的不对称性，是相对于平均值不对称程度的度量，通过对偏度系数的测量，我们能够判定数据分布的不对称程度以及方向。

偏度的衡量是相对于正态分布来说，正态分布的偏度为0，即若数据分布是对称的，偏度为0。若偏度大于0，则分布右偏，即分布有一条长尾在右；若偏度小于0，则分布为左偏，即分布有一条长尾在左（如下图）；同时偏度的绝对值越大，说明分布的偏移程度越严重。

![IwYw9D](https://oss.images.shujudaka.com/uPic/IwYw9D.png)

【注意】数据分布的左偏或右偏，指的是数值拖尾的方向，而不是峰的位置。

* 峰度/Kurtosis

峰度，是研究数据分布陡峭或者平滑的统计量，通过对峰度系数的测量，我们能够判定数据相对于正态分布而言是更陡峭还是更平缓。

描述数据分布形态的陡缓程度（见下图）。当峰度≈0时，可认为分布的峰态合适，服从正态分布（不胖不瘦）；当峰度>0时，分布的峰态陡峭（高尖）；当峰度<0时，分布的峰态平缓（矮胖）。

![kbo6Nb](https://oss.images.shujudaka.com/uPic/kbo6Nb.png)

利用偏度和峰度进行正态性检验时，可以同时计算其相应的Z评分（Z-score），即：偏度Z-score=偏度值/标准误，峰度Z-score=峰度值/标准误。在α=0.05的检验水平下，若Z-score在±1.96之间，则可认为资料服从正态分布。


### 图形判断

直方图、概率图和分位数图都可以直观地看出是否是正态分布，具体见下图。直方图表示连续性变量的频数分布，可以用来考察分布是否服从正态分布。服从正态分布时，图形为倒钟型，如下所示。

![oveDbS](https://oss.images.shujudaka.com/uPic/oveDbS.png)

概率图反映了变量的实际累积概率与理论累积概率的符合程度。分位数图反映了变量的实际分布与理论分布的符合程度。若数据服从正态分布，则数据点应与理论直线（即对角线）基本重合；若数据不服从正态分布，则数据点会偏离很多，如上图所示。

P-P图和Q-Q图都是用来观察变量是否服从正太分布的。

P-P图是根据变量的累积概率对应于所指定的理论分布累积概率绘制的散点图，用于直观地检测样本数据是否符合某一概率分布。如果被检验的数据符合所指定的分布，则代表样本数据的点应当基本在代表理论分布的对角线上。

Q-Q图的结果与P-P图非常相似，只是P-P图是用分布的累计比，而Q-Q图用的是分布的分位数来做检验。和P-P图一样，如果数据为正态分布，则在Q-Q正态分布图中，数据点应基本在图中对角线上。

由于P-P图和Q-Q图的用途完全相同，只是检验方法存在差异。要利用QQ图鉴别样本数据是否近似于正态分布，只需看QQ图上的点是否近似地在一条直线附近，而且该直线的斜率为标准差，截距为均值。

QQ图和PP图的定义方式一样，二者的区别是pp图比较的是真实的数据和待检验分布的累计概率，而qq图比较的是真实数据和待检验分布的分位点数。

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

n = 100
samples = st.norm.rvs(loc = 5, scale = 2, size = n)
samples_sort = sorted(samples)

probplot = sm.ProbPlot(samples, dist = st.norm, loc = 5, scale = 2)
probplot.qqplot(line='45')
plt.show()


x_labels_q = samples_sort
y_labels_q = st.norm.ppf(x_labels_p, loc = 5, scale = 2)
plt.scatter(x_labels_q, y_labels_q)
plt.title('QQ plot for normal distribution samle')
plt.show()

x_labels_p = np.arange(1/(2*n), 1, 1/n)
y_labels_p = st.norm.cdf(samples_sort, loc = 5, scale = 2)
plt.scatter(x_labels_p, y_labels_p)
plt.title('PP plot for normal distribution samle')
plt.show()
```

### 非参数检验

常用的非参数检验方法包括Shapiro-Wilk检验（S-W检验）和Kolmogorov–Smirnov检验（K-S检验）。S-W检验适用于小样本资料，而K-S检验适用于大样本资料。以大家熟悉的SPSS软件为例，当样本量≤5000时，SPSS可以同时报告S-W检验和K-S检验两个结果，但当样本量>5000时，只显示K-S检验结果。

虽然非参数检验方法可以不费脑不费眼的判断是否服从正态分布，但是也具有局限性。当样本量较少的时候，检验结果不够敏感，即使数据分布有一定的偏离也不一定能检验出来；而当样本量较大的时候，检验结果又会太过敏感，只要数据稍微有一点偏离，P值就会<0.05，检验结果倾向于拒绝原假设，认为数据不服从正态分布。

鉴于非参数检验的局限性，不建议完全依赖非参数检验。大家可以平衡多种方法，结合图形法和非参数检验来判断是否服从正态分布。

## 那些被问爆了的典型问题

### 典型问题1

1. 两组数据正态检验，样本量皆>50，但正态性检验K-S和S-W的结果不一致，前者显示非正态，后者显示为正态，请问该以哪个检验结果为准？

2．60人分成3组，每组20人，每个组变量的正态性检验用什么方法最合适？

首选方法是图形观察，即利用直方图、P-P图或Q-Q图进行观察，如果分布严重偏态和尖峰分布则建议进行进一步的假设检验。在样本量较小的情况下（小于50个样本的情况下），请优先选择S-W检验。

### 典型问题2

往往会出现这样的情况，明明直方图显示分布很对称，但正态性检验的结果P 值却<0.05，拒绝原假设认为不服从正态分布，怎么办？

![kGXAtr](https://oss.images.shujudaka.com/uPic/kGXAtr.png)

从实用性的角度，非参数检验不如图形工具进行直观判断好用。非参数检验非常依赖样本量，如果样本量足够多，非常容易得出P<0.05的检验结果。此时并不能说明数据来自的总体不服从正态分布。因此，检验是否服从正态分布时，不能太依赖于非参数检验，一定要参考直方图、P-P图等图形工具来帮助判断。

此外，很多统计学方法，如T检验、方差分析等，与其说要求数据严格服从正态分布，不如说“数据分布不要过于偏态”更为合适。
    

### 内核密度图/Kernel density plots和 Seaborn 简介

到目前为止，您一直只使用 Pandas 中的绘图方法。 现在，您将使用 Seaborn 包。 

Seaborn 是一个较新的 Python 包，它抽象了较低级别的 matplotlib 图表。 Seaborn 还实现了一些额外的尖端图表类型。

**核密度估计**或**kde**图在概念上类似于直方图。 

核密度图显示数据值的平滑密度曲线的值。 换句话说，核密度图是直方图的平滑版本。
 
下面单元格中的代码按照之前使用的方法创建核密度图。 使用 Seaborn 将以下内容添加到配方中：

1. 设置绘图网格的样式。
2. 使用引擎大小列作为参数，使用“distplot”定义绘图类型。 在这种情况下，没有绘制直方图。 显示轴上数据点位置的“地毯”沿水平轴显示。
3. 再次使用 Matplotlib 方法将注释添加到图中。 Seaborn 建立在 Matplotlib 之上，因此始终可以混合使用 Matplotlib 方法。
 
****
**注意：** 根据您的平台和 numpy 版本，您可能会看到弃用警告。 您可以安全地忽略此警告。
****

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
def plot_density_hist(auto_prices, cols, bins = 10, hist = False):
    for col in cols:
        sns.set_style("whitegrid")
        sns.distplot(auto_prices[col], bins = bins, rug=True, hist = hist)
        plt.title('Histogram of ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel('Number of autos')# Set text for y axis
        plt.show()
        
plot_density_hist(auto_prices, num_cols)        
```

kde 图显示了与直方图相同的偏度属性。 rug 显示轴上数据点密度的不同视图。 

一些细节在这个视图中更加明显。

## 二维图

使用汇总统计和几种一维绘图方法探索数据后，您将使用**二维绘图**继续探索。 二维图可帮助您理解**两个变量之间的关系**。 对于机器学习，最感兴趣的关系是**特征**和**标签**之间的关系。 检查特征之间的关系以确定特征是否协变也很有用。 当关系是非线性时，这样的过程可以证明比简单地计算相关性更可靠。

### 创建散点图

散点图广泛用于检查两个变量之间的关系。 在这种情况下，创建的图是一些特征与标签、汽车价格的对比。

下面单元格中的代码遵循之前使用 Pandas 绘图方法的方法，使用 plot.scatter 方法。 此方法有两个必需的 x 轴和 y 轴参数。 执行此代码并检查结果。

```python
def plot_scatter(auto_prices, cols, col_y = 'price'):
    for col in cols:
        fig = plt.figure(figsize=(7,6)) # define plot area
        ax = fig.gca() # define axis   
        auto_prices.plot.scatter(x = col, y = col_y, ax = ax)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col)   # Set text for the x axis
        ax.set_ylabel(col_y) # Set text for y axis
        plt.show()

num_cols = ['curb_weight', 'engine_size', 'horsepower', 'city_mpg']
plot_scatter(auto_prices, num_cols)        
```

这些图显示了这些特征与标签之间的密切关系。 这些特征很可能对预测汽车价格有用。 engine_size 和 horsepower 与价格有相当线性的关系，而 curb_weight 尤其是 city_mpg 则没有。

hoursepower 和 engine_size 似乎是共线的。 要检验这个假设，请执行下面单元格中的代码并检查结果。

```python
plot_scatter(auto_prices, ['horsepower'], 'engine_size') 
```

事实上，这些特征确实呈线性相关。 因此，您不会希望在同一个机器学习模型中使用它们。

### 处理过度绘制/overplotting

检查上面的 engine_size 或 city_mpg 与价格。 请注意，对于某些发动机尺寸/engine sizes和 city_mpg 评级，有许多汽车的价格范围很窄。 

显然，汽车制造商瞄准了这些特性。 结果是许多点在散点图上一个接一个地绘制，导致**过度绘制**。 

当散点图应用于大型数据集时，过度绘制是一个严重的问题。 严重的过度策划会使情节变得毫无意义或无法解释。

幸运的是，有几种处理过度绘制的好方法：

1. 使用点的**透明度**让视图可以看到点。 对于轻微的过度绘制，这种方法可能非常有效。
2. **Countour plots** 或 **2d density plots** 显示点的密度，就像地形图显示高程一样。 生成轮廓具有很高的计算复杂度，使该方法不适用于海量数据集。
3. **Hexbin 图**是直方图的二维模拟。 六边形单元中的阴影密度表示点的密度。 生成 hexbins 在计算上是高效的，可以应用于海量数据集。

下面单元格中的代码修改了之前用于添加透明度参数的散点图函数。 

在统计图形中，alpha，透明度的倒数被指定； 

* alpha = 1.0 是不透明的，
* alpha = 0.0 是完全透明的。 

下面单元格中的代码使用 0.2 的低 alpha（高透明度）。 执行此代码并检查结果。

```python
def plot_scatter_t(auto_prices, cols, col_y = 'price', alpha = 1.0):
    for col in cols:
        fig = plt.figure(figsize=(7,6)) # define plot area
        ax = fig.gca() # define axis   
        auto_prices.plot.scatter(x = col, y = col_y, ax = ax, alpha = alpha)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel(col_y)# Set text for y axis
        plt.show()

plot_scatter_t(auto_prices, num_cols, alpha = 0.2)        
```

凭借高透明度，现在可以查看密集区域中的点以获得更好的数据视图。

使用透明度进行重叠绘制很有用，但也有局限性。 对于大量的点，您将需要其他方法。 使用等值线图或二维密度图就是这样一种解决方案。 

下面单元格中的代码使用了 Seaborn 的 `joinplot` 函数。 该图显示 1d KDE 图以及显示 2d 密度的计数图。 执行此代码并检查结果。

```python
def plot_desity_2d(auto_prices, cols, col_y = 'price', kind ='kde'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.jointplot(col, col_y, data=auto_prices, kind=kind)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()

plot_desity_2d(auto_prices, num_cols)      
```

这些密度等高线图显示了这些特征与标签之间关系的完全不同的观点。 特别是，对于 curb_weight、horsepower，尤其是 city_mpg，二维多模态行为是可见的。

另请注意，显示了相关系数。

下面单元格中的代码显示相同变量的二维 hexbin 图和一维直方图。 执行此代码并检查结果。

```python
plot_desity_2d(auto_prices, num_cols, kind = 'hex')   
```

hexbin 图的总体印象与等高线图大致相同。 由于显示了只有 1 个点的单元格，因此可以看到更多细节。

### 分类变量和数值变量之间的关系

您已经创建了数字变量的二维图但是，如果某些特征是分类变量，您可以做什么？ 有两种专门用于这种情况的绘图类型：

1. **箱形图**突出了分布的四分位数。 毫不奇怪，箱线图包含一个盒子。 **内部两个四分位数**的范围包含在方框内。 方框的长度显示了**四分位数范围**。 方框内的一条线显示中位数。 **Wiskers** 最大延伸至四分位距或数据极值的 1.5 倍。 超出晶须的异常值以符号显示/Outliers beyond the wiskers are shown in a symbol. 。

2. **小提琴图**是一维 KDE 图的变体。 两条背靠背 KDE 曲线用于显示密度估计。

小提琴图的箱线图可以与按分类变量类别分组的数值变量数据并排排列。 以这种方式，每个盒子或小提琴显示代表分类变量的每个类别的个案的数值变量的值。

执行下面单元格中的代码以显示分类变量列表的箱线图，并检查结果。

```python
def plot_box(auto_prices, cols, col_y = 'price'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot(col, col_y, data=auto_prices)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
cat_cols = ['fuel_type', 'aspiration', 'num_of_doors', 'body_style', 
            'drive_wheels', 'engine_location', 'engine_type', 'num_of_cylinders']
plot_box(auto_prices, cat_cols)    
```

对于每个分类变量，您可以看到为每个唯一类别创建了一个箱线图。 请注意，对于其中一些情况，按类别划分的汽车价格之间存在一些明显差异。 例如，对于 fuel_type 或 aspiration，存在明显差异。 在其他情况下，例如 num_of_doors，差异并不显着。 对于 num_of_cylinders 有显着差异，但有两个类别只有一个案例，这是有问题的。

下面单元格中的代码使用小提琴图创建与上面类似的显示。 执行代码并检查结果。

```python
def plot_violin(auto_prices, cols, col_y = 'price'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.violinplot(col, col_y, data=auto_prices)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
plot_violin(auto_prices, cat_cols)    
```

小提琴图的解释类似于箱线图。 

但是，可以看到有关分布的更多详细信息。 每个小提琴的面积在每个绘图显示上都是相同的。 另请注意，在每个小提琴图中都可以看到一种箱线图。

当您检查上面的图时，请注意某些关系更为明显。 

例如，很明显，车门的数量不会影响汽车的价格，而车身风格最多只能产生边际影响。 

而 engine_location 和 num_of_cylinders 确实会影响价格。

## 使用aesthetics/美学来添加项目额外的维度

到目前为止，您只能在一个图上使用一个或两个变量。 

但是，对于复杂的数据集，在每个图上查看多个维度很有用。 

问题是，当图形显示仅限于二维时，如何做到这一点？

在本节中，plot aesthetics 用于投射额外的维度。 一些aesthetics仅对分类变量有用，而其他aesthetics对数值变量有用。 

请记住，并非所有的plot aesthetics都同样有效。 

对人类感知的测试表明，人们非常善于注意到位置上的微小差异。 这个事实解释了为什么散点图如此有效。 按照有效性的粗略顺序，这些aesthetics是：


1. **Marker shape/标记形状** 是一个有效的指标变量类别。 选择易于被观察者区分的形状是至关重要的。
2. **Marker size/标记大小** 显示数值变量的值。 小心，因为大小是跨标记的跨度，而不是区域。
3. **Marker color/标记颜色** 可用作变量类别的指示器。 就人类感知而言，颜色是这三种aesthetics/美学中最不有效的。 应选择颜色以显得不同。 此外，请记住，很多人，尤其是男性，都是红绿色盲。

Categorical aesthetics/分类美学，例如标记的形状和颜色，只有在标记的差异是可感知的情况下才有效。 

使用太多的形状或颜色会导致观众无法分辨类别之间的差异。 

**通常应遵守大约五到七个类别的限制。**

### 标记形状

下面单元格中的代码使用标记形状在散点图上显示汽车的燃料类型。 这是通过用分类列的每个唯一值对数据框进行子集化来完成的。 形状在列表中定义，在此内部循环的每次迭代中都会引用该列表。

这段代码有一个棘手的方面。 透明度参数 alpha 必须通过字典传递给 Matplotlib。 键是参数 alpha，值是参数值。

执行此代码并检查结果。

```python
def plot_scatter_shape(auto_prices, cols, shape_col = 'fuel_type', col_y = 'price', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    unique_cats = auto_prices[shape_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            temp = auto_prices[auto_prices[shape_col] == cat]
            sns.regplot(col, col_y, data=temp, marker = shapes[i], label = cat,
                        scatter_kws={"alpha":alpha}, fit_reg = False, color = 'blue')
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()
            
num_cols = ['curb_weight', 'engine_size', 'horsepower', 'city_mpg']
plot_scatter_shape(auto_prices, num_cols)                
```

虽然存在一些重叠，但汽油和柴油汽车之间的差异现在在这些图中很明显。 这种新的数据视图有助于确认燃料类型是确定汽车价格的重要特征。

请注意，已为该显示选择了相当独特的形状。 总之，通过按类别添加形状，第三维被投影到这些图上。

### 标记大小

下面单元格中的代码使用标记大小来显示整备重量。 由于 Matplotlib 使用面积来计算标记大小，因此整备重量的值被平方，然后通过一个方便的乘数进行缩放。 请注意，由于大小 `s` 是 Matplotlib 参数，因此它与 alpha 一起在字典中传递。 执行此代码并检查结果。

```python
def plot_scatter_size(auto_prices, cols, shape_col = 'fuel_type', size_col = 'curb_weight',
                            size_mul = 0.000025, col_y = 'price', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    unique_cats = auto_prices[shape_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            temp = auto_prices[auto_prices[shape_col] == cat]
            sns.regplot(col, col_y, data=temp, marker = shapes[i], label = cat,
                        scatter_kws={"alpha":alpha, "s":size_mul*temp[size_col]**2}, 
                        fit_reg = False, color = 'blue')
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()

num_cols = ['engine_size', 'horsepower', 'city_mpg']
plot_scatter_size(auto_prices, num_cols)  
```

这些图有几个有趣的方面，每个方面都有助于预测汽车的价格 对于柴油汽车，curb_weight、价格、engine_size、马力和 city_mpg 之间的关系很复杂，没有明确的趋势。 另一方面，价格高、发动机大、马力大、city_mpg 低的汽车似乎配备大型汽油发动机。

上面的图现在在 2d 图表面上投影了四个维度。

### 颜色

正如已经讨论过的，颜色的变化对许多人来说很难察觉。 尽管如此，颜色对于投影有限数量的变量类别还是很有用的。 选择独特的颜色有助于这种情况。

下面单元格中的代码使用颜色来显示汽车的吸气类别。 两个内部循环创建数据子集，这些子集使用特定形状和颜色的标记绘制。 执行此代码并检查结果。

```python
def plot_scatter_shape_size_col(auto_prices, cols, shape_col = 'fuel_type', size_col = 'curb_weight',
                            size_mul = 0.000025, color_col = 'aspiration', col_y = 'price', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    colors = ['green', 'blue', 'orange', 'magenta', 'gray'] # specify distinctive colors
    unique_cats = auto_prices[shape_col].unique()
    unique_colors = auto_prices[color_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            for j, color in enumerate(unique_colors):
                temp = auto_prices[(auto_prices[shape_col] == cat) & (auto_prices[color_col] == color)]
                sns.regplot(col, col_y, data=temp, marker = shapes[i],
                            scatter_kws={"alpha":alpha, "s":size_mul*temp[size_col]**2}, 
                            label = (cat + ' and ' + color), fit_reg = False, color = colors[j])
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()

num_cols = ['engine_size', 'horsepower', 'city_mpg']        
plot_scatter_shape_size_col(auto_prices, num_cols)     
```

这些图中的每一个都将五个维度的数据投影到二维显示器上。 现在这些数据中有几种关系是显而易见的。

总之，aspiration 和 fuel_type 应该是有用的价格预测指标。

现在，回答课程页面上的**问题 3**。

颜色（或色调）可用于其他类型的绘图。 例如，下面单元格中的代码显示拆分的小提琴图。 小提琴按类型划分，并以不同颜色显示。 执行此代码并检查结果。

```python
def plot_violin_hue(auto_prices, cols, col_y = 'price', hue_col = 'aspiration'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.violinplot(col, col_y, data=auto_prices, hue = hue_col, split = True)
        plt.xlabel(col)  # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
plot_violin_hue(auto_prices, cat_cols)    
```

这些图表明，配备涡轮增压发动机/turbo aspiration的汽车通常比同类标准汽车更贵。 因此，aspiration 应该是一个有用的价格预测指标。

## 数据的多轴视图

到目前为止，您一直在处理具有一对轴的图。 但是，很有可能创建具有多个轴的强大数据可视化。 

这些方法允许您在一个视图中检查多个变量之间的关系。 这些多个视图有助于理解复杂数据集中的许多关系。 有许多强大的多轴绘图方法。 

在本实验中，您将使用两种常用方法：

1. **成对散点图**或**散点图矩阵**是一组散点图，其公共轴沿数组的行和列。 数组的对角线可用于显示分布图。 对角线上方或下方的单元格可用于其他绘图类型，如等高线密度图。

2. **条件图**、**多面图**或**小多图**使用**分组依据**操作来创建和显示数据集的子集。 显示可以是按数据集分组组织的一维或二维数组。

### 成对散点图

您现在将向 auto.price 数据集应用散点图矩阵。 下面单元格中的代码使用了 Seaborn 包中的“pairplot”函数。 此函数在对角线下方创建一个基本散点图矩阵。 每个变量的核密度估计显示在对角线上。 使用“map_upper”方法，二维密度图显示在对角线上方。 运行下面的单元格以创建数据集中数字特征的散点图矩阵。

```python
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
num_cols = ["curb_weight", "engine_size", "horsepower", "city_mpg", "price", "fuel_type"] 
sns.pairplot(auto_prices[num_cols], 
             hue='fuel_type', 
             palette="Set2", 
             diag_kind="kde", 
             size=2).map_upper(sns.kdeplot, cmap="Blues_d")
```

查看散点图矩阵（如果该图对于单元格而言太大，您可以通过单击其左边距来展开单元格）。

请注意，该图由许多散点图组成。 对于每个变量，都有一行和一列。 变量绘制在行的垂直轴和列的水平轴上。 这样，所有变量的交叉图的每个组合都以两种可能的方向显示。 每个变量的 KDE 图都在对角线上。 在对角线上方，您可以看到二维密度估计的等值线图。 这里有很多细节。

### 条件图/Conditioned plots

现在您将探索条件图的使用。 下面单元格中的代码执行以下操作：

1. Seaborn 的“FacetGrid”函数定义了显示绘图的网格对象。 此函数的参数是 Pandas 数据框以及行和列的分组变量。
2. `map` 方法在绘图网格上显示（映射）绘图函数。 在本例中，使用了来自 Matplotlib 的直方图函数“hist”。

执行代码并检查结果。

```python
## Function to plot conditioned histograms
def cond_hists(df, plot_cols, grid_col):
    import matplotlib.pyplot as plt
    import seaborn as sns
    ## Loop over the list of columns
    for col in plot_cols:
        grid1 = sns.FacetGrid(df, col=grid_col)
        grid1.map(plt.hist, col, alpha=.7)
    return grid_col

## Define columns for making a conditioned histogram
plot_cols2 = ["length",
               "curb_weight",
               "engine_size",
               "city_mpg",
               "price"]

cond_hists(auto_prices, plot_cols2, 'drive_wheels')
```

检查这一系列条件图。 以 drive_wheels 类别为条件的数值特征的分布存在一致差异。

现在，回答课程页面上的**问题 4**。

接下来，您将创建和检查条件散点图。 通过调节多个维度，您可以将几个额外的维度投影到二维图上。 调节可以被认为是分组操作。

在本练习中，您不会使用点形状作为区分因素，但请记住，形状与颜色一样有用。 此外，对于很大一部分色盲人群来说，形状可能更容易。

注意：组合投影多个维度的方法时要小心。 你很容易得到一个不仅难以解释，而且更难将你的观察结果传达给同事的情节。 例如，如果您使用三个条件变量，再加上颜色和形状，那么您将投射数据集的七个维度。 虽然这种方法可能会揭示重要的关系，但它可能只会创建一个复杂的情节。

下面单元格中的代码使用 Python seaborn 包按以下方式创建散点图的二维数组：

1. 和前面一样，定义网格对象。 颜色美学与指定的调色板一起使用。
2. 散点图映射到网格。

执行此代码并检查结果。

```python
def cond_plot(cols):
#     import IPython.html.widgets
    import seaborn as sns
    for col in cols:
        g = sns.FacetGrid(auto_prices, col="drive_wheels", row = 'body_style', 
                      hue="fuel_type", palette="Set2", margin_titles=True)
        g.map(sns.regplot, col, "price", fit_reg = False)

num_cols = ["curb_weight", "engine_size", "city_mpg"]
cond_plot(num_cols)    
```

仔细检查您创建的图。 这些图总共显示了数据的五个维度，并且有很多细节需要理解。 通常，您应该能够得出以下结论：

- 对于某些条件变量组合，根本没有汽车。 例如，没有四轮驱动敞篷车或硬顶。 因此，这些地块是空白的。
- 某些空调组合没有柴油车，例如敞篷车和四轮驱动车。
- 有许多不同的分组。 例如，前轮驱动和后轮驱动的轿车或货车具有截然不同的行为。

同样，许多这些观察结果可以用其他类型的图进行。 但是，将这些图布置在数据集的子组中是有好处的。 数据的不同子分组突出了数据的不同关系。
