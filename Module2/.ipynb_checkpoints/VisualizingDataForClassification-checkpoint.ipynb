# 可视化分类数据

在上一个实验中，您探索了汽车价格数据集以了解回归问题的关系。 在本实验中，您将探索德国银行信贷数据集以了解**分类**问题的关系。 

不同之处在于，在分类问题中，标签是分类变量。

在其他实验室中，您将使用通过可视化学到的知识来创建一个解决方案来预测信用不佳的客户。 

目前，该实验室的重点是直观地探索数据以确定哪些特征可能有助于预测客户的不良信用。

分类问题的可视化与回归问题的可视化有很多共同之处。 应识别共线特征，以便可以消除或以其他方式处理它们。 但是，对于分类问题，您正在寻找有助于**分离标签类别**的功能。 当每个标签类别都有不同的特征值时，就实现了分离。 良好的分离导致低分类错误率。

## 加载并准备数据集

作为第一步，您必须加载数据集。

执行下面单元格中的代码以加载本笔记本其余部分所需的包。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy.random as nr
import math

%matplotlib inline
```

下面单元格中的代码加载数据集并将人类可读的名称分配给列。 然后打印数据框的形状和头部。 执行这段代码：

```python
credit = pd.read_csv('German_Credit.csv', header=None)
credit.columns = ['customer_id',
                  'checking_account_status', 'loan_duration_mo', 'credit_history', 
                  'purpose', 'loan_amount', 'savings_account_balance', 
                  'time_employed_yrs', 'payment_pcnt_income','gender_status', 
                  'other_signators', 'time_in_residence', 'property', 'age_yrs',
                  'other_credit_outstanding', 'home_ownership', 'number_loans', 
                  'job_category', 'dependents', 'telephone', 'foreign_worker', 
                  'bad_credit']
print(credit.shape)
credit.head()
```

数据集中有 1011 行和 22 列。 第一列是 customer_id，这是一个标识符。 我们将放弃它，因为这不是一项功能。

```python
credit.drop(['customer_id'], axis=1, inplace=True)
print(credit.shape)
credit.head()
```

现在，还剩下 21 列。 在 21 列中，有 20 个特征加上一个标签列。 

这些特征代表银行可能拥有的关于其客户的信息。 有数字和分类特征。 然而，分类特征的编码方式使它们难以理解。 

此外，标签被编码为 $\{ 1,2 \}$ 这有点尴尬。

下面单元格中的代码使用字典列表重新编码具有人类可读文本的分类特征。 

列表中的最后一个字典将良好和不良信用重新编码为二进制变量 $\{ 0,1 \}$。 

`for` 循环遍历列并将代码映射到人类可读的类别名称。 拥有人类可读的数据编码极大地提高了人们理解数据中关系的能力。

执行这段代码并检查结果：

```python
code_list = [['checking_account_status', 
              {'A11' : '< 0 DM', 
               'A12' : '0 - 200 DM', 
               'A13' : '> 200 DM or salary assignment', 
               'A14' : 'none'}],
            ['credit_history',
            {'A30' : 'no credit - paid', 
             'A31' : 'all loans at bank paid', 
             'A32' : 'current loans paid', 
             'A33' : 'past payment delays', 
             'A34' : 'critical account - other non-bank loans'}],
            ['purpose',
            {'A40' : 'car (new)', 
             'A41' : 'car (used)',
             'A42' : 'furniture/equipment',
             'A43' : 'radio/television', 
             'A44' : 'domestic appliances', 
             'A45' : 'repairs', 
             'A46' : 'education', 
             'A47' : 'vacation',
             'A48' : 'retraining',
             'A49' : 'business', 
             'A410' : 'other' }],
            ['savings_account_balance',
            {'A61' : '< 100 DM', 
             'A62' : '100 - 500 DM', 
             'A63' : '500 - 1000 DM', 
             'A64' : '>= 1000 DM',
             'A65' : 'unknown/none' }],
            ['time_employed_yrs',
            {'A71' : 'unemployed',
             'A72' : '< 1 year', 
             'A73' : '1 - 4 years', 
             'A74' : '4 - 7 years', 
             'A75' : '>= 7 years'}],
            ['gender_status',
            {'A91' : 'male-divorced/separated', 
             'A92' : 'female-divorced/separated/married',
             'A93' : 'male-single', 
             'A94' : 'male-married/widowed', 
             'A95' : 'female-single'}],
            ['other_signators',
            {'A101' : 'none', 
             'A102' : 'co-applicant', 
             'A103' : 'guarantor'}],
            ['property',
            {'A121' : 'real estate',
             'A122' : 'building society savings/life insurance', 
             'A123' : 'car or other',
             'A124' : 'unknown-none' }],
            ['other_credit_outstanding',
            {'A141' : 'bank', 
             'A142' : 'stores', 
             'A143' : 'none'}],
             ['home_ownership',
            {'A151' : 'rent', 
             'A152' : 'own', 
             'A153' : 'for free'}],
            ['job_category',
            {'A171' : 'unemployed-unskilled-non-resident', 
             'A172' : 'unskilled-resident', 
             'A173' : 'skilled',
             'A174' : 'highly skilled'}],
            ['telephone', 
            {'A191' : 'none', 
             'A192' : 'yes'}],
            ['foreign_worker',
            {'A201' : 'yes', 
             'A202' : 'no'}],
            ['bad_credit',
            {2 : 1,
             1 : 0}]]

for col_dic in code_list:
    col = col_dic[0]
    dic = col_dic[1]
    credit[col] = [dic[x] for x in credit[col]]
    
credit.head()    
```

分类特征现在具有有意义的编码。 此外，标签现在被编码为二进制变量。

## 检查类和类不平衡/Examine classes and class imbalance

在这种情况下，标签具有显着的**类别不平衡**。 

类别不平衡意味着标签类别的案例数量不相等。 

类不平衡会严重影响分类器算法的训练。 

在许多情况下，这种不平衡会导致少数类的错误率更高。 

大多数现实世界的分类问题都存在类不平衡，有时甚至是严重的类不平衡，因此在训练任何模型之前对此进行测试很重要。

幸运的是，使用频率表很容易测试类不平衡。

执行下面单元格中的代码以显示类别的频率表：

```python
credit_counts = credit['bad_credit'].value_counts()
print(credit_counts)
```

请注意，只有 30% 的案例有不良信用。

这并不奇怪，因为银行通常会留住信用良好的客户。 虽然这不是严重不平衡的情况，但足以使任何模型的训练产生偏差。

## 通过数字特征可视化类分离

如前所述，分类问题可视化的主要目标是了解哪些特征对分类有用。 在本节中，您将从可视化数字特征的分离质量开始。

执行代码，检查结果，并回答课程页面上的**问题 1**。

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
def plot_box(credit, cols, col_x = 'bad_credit'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot(col_x, col, data=credit)
        plt.xlabel(col_x) # Set text for the x axis
        plt.ylabel(col)# Set text for y axis
        plt.show()

num_cols = ['loan_duration_mo', 'loan_amount', 'payment_pcnt_income',
            'age_yrs', 'number_loans', 'dependents']
plot_box(credit, num_cols)
```

您如何解释这些结果？ 

箱线图很有用，因为通过它们的构造，您不得不关注分布四分位数的重叠（或不重叠）。 

在这种情况下，问题是四分位数是否足以使该特征在分离标签类别时有用？ 上图中显示了以下情况：

1. 对于 loan_duration_mo、loan_amount 和 payment as a percent of income (payment_pcnt_income)，有用区分信用良好和不良的客户。 正如人们所预料的那样，信用不良的客户在大额贷款上的贷款期限更长，而且还款额占他们收入的比例更大。

2. 另一方面，age in years、number_loans 和 dependents 似乎并不重要。 在后两种情况下，这种情况似乎是由于中值为零造成的。 只是没有足够的非零案例来制作这些有用的功能。

作为箱线图的替代方法，您可以使用小提琴图来检查按数字特征划分的标签案例。

执行下面单元格中的代码并检查结果：

```python
def plot_violin(credit, cols, col_x = 'bad_credit'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.violinplot(col_x, col, data=credit)
        plt.xlabel(col_x) # Set text for the x axis
        plt.ylabel(col)# Set text for y axis
        plt.show()

plot_violin(credit, num_cols)
```

这些图的解释在很大程度上与箱形图相同。 不过，有一个细节值得注意。 信用良好和不良客户的 loan_duration_mo 和 loan_amount 之间的差异仅针对更极端的值。 可能是这些功能不太有用，箱形图表明了这一点。

## 通过分类特征可视化类分离

现在您将转向可视化分类特征分离标签类别的能力的问题。 理想情况下，分类特征的每个标签值的类别计数都非常不同。 可视化这些关系的一个好方法是使用条形图。

下面单元格中的代码为每个标签类别创建了分类变量的并排图。

执行此代码，检查结果，并回答课程页面上的**问题 2**。

```python
import numpy as np
cat_cols = ['checking_account_status', 'credit_history', 'purpose', 'savings_account_balance', 
                  'time_employed_yrs', 'gender_status', 'other_signators', 'property', 
                  'other_credit_outstanding', 'home_ownership', 'job_category', 'telephone', 
                  'foreign_worker']

credit['dummy'] = np.ones(shape = credit.shape[0])
for col in cat_cols:
    print(col)
    counts = credit[['dummy', 'bad_credit', col]].groupby(['bad_credit', col], as_index = False).count()
    temp = counts[counts['bad_credit'] == 0][[col, 'dummy']]
    _ = plt.figure(figsize = (10,4))
    plt.subplot(1, 2, 1)
    temp = counts[counts['bad_credit'] == 0][[col, 'dummy']]
    plt.bar(temp[col], temp.dummy)
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n Bad credit')
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    temp = counts[counts['bad_credit'] == 1][[col, 'dummy']]
    plt.bar(temp[col], temp.dummy)
    plt.xticks(rotation=90)
    plt.title('Counts for ' + col + '\n Good credit')
    plt.ylabel('count')
    plt.show()
```

这些情节中有很多信息。

解释这些图的关键是比较每个标签值的类别比例。 

如果每个标签类别的这些比例明显不同，则该特征可能有助于分离标签。

这些图中有几种情况很明显：

1. checking_account_status 和credit_history 等一些特征在标签类别之间的类别分布存在显着差异。
2. gender_status 和 telephone 等其他特征差异很小，但这些差异不太可能显着。
3. other_signators、foreign_worker、home_ownership 和 job_category 等其他特征具有主导类别，其他类别的情况很少。 这些特征可能无法区分案例。

请注意，这些分类特征中只有少数对分离个案有用。
